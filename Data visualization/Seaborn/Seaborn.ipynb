## 2. CREATING GRAPHICS WITH SEABORN

### **2.1 Line**

```python
person = ['Maria', 'Carlos', 'Juan', 'Pablo', 'Ana']
salary = [1500, 2300, 3500, 4200, 5100]

dfsalary = pd.DataFrame({'Person' : person,
                         'Salary' : salary})

sns.lineplot(data = dfsalary,  x = 'Person', y = 'Salary')

plt.show()
```

```python
person = ['Maria', 'Carlos', 'Juan', 'Pablo', 'Ana']
salary = [1500, 2300, 3500, 4200, 5100]

dfsalary = pd.DataFrame({'Person' : person,
                         'Salary' : salary})

sns.lineplot(data = dfsalary,  x = 'Person', y = 'Salary', marker = 'o')

plt.show()
```

```python
person = ['Maria', 'Carlos', 'Juan', 'Pablo', 'Ana']
salary = [1500, 2300, 3500, 4200, 5100]
salary_datascience = [3000, 4600, 7000, 8400, 10200]

dfsalary = pd.DataFrame({'Person' : person,
                         'Salary' : salary,
                         'Salary_DataScience' : salary_datascience})

sns.lineplot(data = dfsalary,  x = 'Person', y = 'Salary')
sns.lineplot(data = dfsalary,  x = 'Person', y = 'Salary_DataScience')

plt.show()
```
