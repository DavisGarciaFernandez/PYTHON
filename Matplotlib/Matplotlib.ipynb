# **2. CREATING MATPLOTLIB PLOTS**

## 2.1 Line Graph

```python
person = ['Maria', 'Carlos', 'Juan', 'Pablo', 'Ana']
salary = [1500, 2300, 3500, 4200, 5100]

plt.plot(person, salary)
plt.show()
```

With the plt.plot() function we can draw line graphs.

We can see that on the x and y axes, we have drawn values ​​of the numbers from the lists that have been entered.

We also note that the default color is blue.

```python
person = ['Maria', 'Carlos', 'Juan', 'Pablo', 'Ana']
salary = [1500, 2300, 3500, 4200, 5100]
salary_datascience = [3000, 4600, 7000, 8400, 10200]

plt.plot(person, salary)
plt.plot(person, salary_datascience)

plt.legend(['Salary', 'Salary Data Scientist'])

plt.show()
```

Now when we place another **plt.plot()** with the same X axis but a different Y axis, we obtain the plot of a second line.

This will happen the more **plt.plot()** we place with the same X.

## **2.2 Dot plot**

```python
plt.scatter(person, salary_datascience)
plt.show()
```

Similar to the previous case, just enter both series or list of numerical values ​​in the **plt.scatter()** function to obtain the graph.

Remember that we must always place **plt.show()** to avoid the console output and not the graphics engine.

```python
plt.scatter(person, salary_datascience, label = 'Salary Data science')
plt.scatter(person, salary, label = 'Salary')
plt.legend()
plt.show()
```

By adding another dot plot we can generate a legend where we can specify the difference of the data.

## **2.3 Bar graphic**

For this graph we are going to use the data from the dataset that we loaded at the beginning of the notebook.

Bar charts allow us to see counts, percentages, trends, shares, etc.

The first step is to have the dataset ready to be plotted, for this we will rely on obtaining the frequency of penguins per island.

```python
df.value_counts(['island']).reset_index(name = 'Count')
```

```python
island = ['Biscoe', 'Dream', 'Togersen']
count = [168, 124, 52]

plt.bar(island, count)
plt.show()
```

To avoid writing the list, we can directly call the data column of the dataframe to do the plotting.

```python
dfIslCoun = (
    df.value_counts(['island']).reset_index(name = 'Count')
)

plt.bar(dfIslCoun['island'], dfIslCoun['Count'], color = '#2a9d8f')

plt.show()
```

```python
df_prop_island_species = (
    df
    .value_counts(['island', 'species'])
    .reset_index(name = 'count')
    .sort_values(['island'])
    .assign(prop = lambda x : x.groupby(['island'])['count'].transform(lambda y : y / sum(y)))
)
df_prop_island_species
```

```python
df_prop_island_species.pivot(index ='island', columns = 'species', values = 'prop').plot(kind = 'bar', stacked = True)
plt.show()
```

```python
df_prop_island_species
```

```python
(
 df_prop_island_species
 .drop(columns = ['count'])
 .pivot_table(index = 'species', columns = 'island', values = 'prop', fill_value = 0)
)
```

```python
islands = list(df_prop_island_species['island'].unique())

dictprop = (
    df_prop_island_species
    .drop(columns = ['count'])
    .pivot_table(index = 'island', columns = 'species', values = 'prop', fill_value = 0)
    .reset_index(drop = True)
    .to_dict(orient = 'list')
)

dictprop
```

```python
bar_floot = np.zeros(3)

fig, ax = plt.subplots()

for specie, proportion in dictprop.items():
  ax.bar(
      island,
      proportion,
      label = specie,
      bottom = bar_floot
  )

  bar_floot += proportion

#ax.set_ylim(top = 1.2)
plt.legend()
plt.show()
```

## **2.4 Histograma**

```python
plt.hist(df['bill_depth_mm'])
plt.show()
```

```python
from scipy.stats import gaussian_kde

density = gaussian_kde(df['bill_depth_mm'].dropna())
range_axis_x = np.linspace(
    min(df['bill_depth_mm']),
    max(df['bill_depth_mm'])
)

values_axis_y = density(range_axis_x)

plt.hist(df['bill_depth_mm'], density = True)
plt.plot(range_axis_x, values_axis_y)

plt.show()

# np.linspace(min(data["bill_depth_mm"]), max(data["bill_depth_mm"]))
```

```python
plt.hist(df.query("sex == 'female'")['bill_depth_mm'], label = 'female', alpha = 0.7)
plt.hist(df.query("sex == 'male'")['bill_depth_mm'], label = 'male', alpha = 0.7)
plt.legend()

plt.show()
```
